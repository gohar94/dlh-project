```python
import pandas as pd
import numpy as np
from loaders import *

from scripts.utils import *
```

```python
alexnet_layers = [
    ConfigRegistry.ALEXNET_LAYER1,
    ConfigRegistry.ALEXNET_LAYER2,
    ConfigRegistry.ALEXNET_LAYER3,
    ConfigRegistry.ALEXNET_LAYER4,
    ConfigRegistry.ALEXNET_LAYER5,
]
```

```python
eyeriss_accelergy_result = run_accelergy(
    ConfigRegistry.EYERISS_ARCH,
    ConfigRegistry.EYERISS_COMPONENTS_DIR
)
print(eyeriss_accelergy_result.ert_verbose)
```

```python
alexnet_stats = []
alexnet_loops = []

for i, layer in enumerate(alexnet_layers):
    results = run_timeloop_mapper(
        ConfigRegistry.EYERISS_ARCH,
        ConfigRegistry.EYERISS_COMPONENTS_DIR,
        eyeriss_accelergy_result.art,
        eyeriss_accelergy_result.ert,
        ConfigRegistry.EYERISS_ARCH_CONSTRAINTS,
        ConfigRegistry.EYERISS_MAP_CONSTRAINTS,
        ConfigRegistry.EYERISS_MAPPER,
        layer,
    )
    print(f"Layer: {i+1}")
    stats, loops = results
    print(loops)
    
    alexnet_stats.append(stats)
    alexnet_loops.append(loops)
```

```python
fname = './results/alexnet_eyeriss/mapper/'

extract_numbers(alexnet_stats, fname + 'stats.pkl')
_ = load_data(fname + 'stats.pkl')

save_loop_nests(alexnet_loops, fname + 'loops.txt')
```

```python
for i, (stats, loops) in enumerate(alexnet_eyeriss_layerwise_results):
    print(f"Layer: {i+1}")
    log = False
    for line in stats.split("\n"):
        if "Summary Stats" in line:
            log = True
        if log:
            print(line)
```

```python
total_cycles = 0
for i, (stats, loops) in enumerate(alexnet_eyeriss_layerwise_results):
    for line in stats.split("\n"):
        if "Cycles: " in line:
            cycles = int(line[len("Cycles: "):])
            total_cycles += cycles
            print(f"Cycles in layer {i+1}: {cycles}")
            break
print(f"Eyeriss total cycles: {total_cycles}")
```

```python
total_energy = 0
for i, (stats, loops) in enumerate(alexnet_eyeriss_layerwise_results):
    for line in stats.split("\n"):
        if "Energy: " in line:
            energy = float(line[len("Energy: "):].split(' uJ')[0])
            total_energy += energy
            print(f"Energy in layer {i+1}: {energy}")
            break
print(f"Total energy: {total_energy} uJ")
```
