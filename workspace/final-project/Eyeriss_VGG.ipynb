```python
import pandas as pd
import numpy as np
from loaders import *

from scripts.utils import *
```

```python
vgg_layers = [
    ConfigRegistry.VGG_LAYER1,
    ConfigRegistry.VGG_LAYER2,
    ConfigRegistry.VGG_LAYER3,
    ConfigRegistry.VGG_LAYER4,
    ConfigRegistry.VGG_LAYER5,
    ConfigRegistry.VGG_LAYER6,
    ConfigRegistry.VGG_LAYER7,
    ConfigRegistry.VGG_LAYER8,
    ConfigRegistry.VGG_LAYER9,
    ConfigRegistry.VGG_LAYER10,
    ConfigRegistry.VGG_LAYER11,
    ConfigRegistry.VGG_LAYER12,
    ConfigRegistry.VGG_LAYER13
]
```

```python
eyeriss_accelergy_result = run_accelergy(
    ConfigRegistry.EYERISS_ARCH,
    ConfigRegistry.EYERISS_COMPONENTS_DIR
)
print(eyeriss_accelergy_result.ert_verbose)
```

```python
vgg16_stats = []
vgg16_loops = []
vgg16_eyeriss_layerwise_results = []

for i, layer in enumerate(vgg_layers):
    results = run_timeloop_mapper(
        ConfigRegistry.EYERISS_ARCH,
        ConfigRegistry.EYERISS_COMPONENTS_DIR,
        eyeriss_accelergy_result.art,
        eyeriss_accelergy_result.ert,
        ConfigRegistry.EYERISS_ARCH_CONSTRAINTS,
        ConfigRegistry.EYERISS_MAP_CONSTRAINTS,
        ConfigRegistry.EYERISS_MAPPER,
        layer,
    )
    print(f"Layer: {i+1}")
    stats, loops = results
    print(loops)
    
    vgg16_stats.append(stats)
    vgg16_loops.append(loops)
    vgg16_eyeriss_layerwise_results.append(results)
```

```python
for i, (stats, loops) in enumerate(vgg16_eyeriss_layerwise_results):
    print(f"Layer: {i+1}")
    log = False
    for line in stats.split("\n"):
        if "Summary Stats" in line:
            log = True
        if log:
            print(line)
```

```python
total_cycles = 0
for i, (stats, loops) in enumerate(vgg16_eyeriss_layerwise_results):
    for line in stats.split("\n"):
        if "Cycles: " in line:
            cycles = int(line[len("Cycles: "):])
            total_cycles += cycles
            print(f"Cycles in layer {i+1}: {cycles}")
            break
print(f"Eyeriss total cycles: {total_cycles}")
```

```python
total_energy = 0
for i, (stats, loops) in enumerate(vgg16_eyeriss_layerwise_results):
    for line in stats.split("\n"):
        if "Energy: " in line:
            energy = float(line[len("Energy: "):].split(' uJ')[0])
            total_energy += energy
            print(f"Energy in layer {i+1}: {energy}")
            break
print(f"Total energy: {total_energy} uJ")
```
