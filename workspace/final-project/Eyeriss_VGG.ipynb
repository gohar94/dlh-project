```python
import pandas as pd
import numpy as np
from loaders import *
```

```python
vgg_layers = [
    ConfigRegistry.VGG01_LAYER1,
    ConfigRegistry.VGG01_LAYER2,
    ConfigRegistry.VGG01_LAYER3,
    ConfigRegistry.VGG01_LAYER4,
    ConfigRegistry.VGG01_LAYER5,
    ConfigRegistry.VGG01_LAYER6,
    ConfigRegistry.VGG01_LAYER7,
    ConfigRegistry.VGG01_LAYER8,
]
```

```python
eyeriss_accelergy_result = run_accelergy(
    ConfigRegistry.EYERISS_ARCH,
    ConfigRegistry.EYERISS_COMPONENTS_DIR
)
print(eyeriss_accelergy_result.ert_verbose)
```

```python
vgg_eyeriss_layerwise_results = []

for i, layer in enumerate(vgg_layers):
    eyeriss_results = run_timeloop_mapper(
        ConfigRegistry.EYERISS_ARCH,
        ConfigRegistry.EYERISS_COMPONENTS_DIR,
        eyeriss_accelergy_result.art,
        eyeriss_accelergy_result.ert,
        ConfigRegistry.EYERISS_ARCH_CONSTRAINTS,
        ConfigRegistry.EYERISS_MAP_CONSTRAINTS,
        ConfigRegistry.EYERISS_MAPPER,
        layer,
    )
    print(f"Layer: {i+1}")
    eyeriss_results_stats, eyeriss_results_loops = eyeriss_results
    print(eyeriss_results_loops)
    print(eyeriss_results_stats)
    vgg_eyeriss_layerwise_results.append(eyeriss_results)
```

```python
for i, (stats, loops) in enumerate(vgg_eyeriss_layerwise_results):
    print(f"Layer: {i+1}")
    log = False
    for line in stats.split("\n"):
        if "Summary Stats" in line:
            log = True
        if log:
            print(line)
```

```python
total_cycles = 0
for i, (stats, loops) in enumerate(vgg_eyeriss_layerwise_results):
    for line in stats.split("\n"):
        if "Cycles: " in line:
            cycles = int(line[len("Cycles: "):])
            total_cycles += cycles
            print(f"Cycles in layer {i+1}: {cycles}")
            break
print(f"Eyeriss total cycles: {total_cycles}")
```

```python
total_energy = 0
for i, (stats, loops) in enumerate(vgg_eyeriss_layerwise_results):
    for line in stats.split("\n"):
        if "Energy: " in line:
            energy = float(line[len("Energy: "):].split(' uJ')[0])
            total_energy += energy
            print(f"Energy in layer {i+1}: {energy}")
            break
print(f"Total energy: {total_energy} uJ")
```
